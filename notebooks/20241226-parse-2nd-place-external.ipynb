# Excute the capypara.py, rlhflow1.py, rlhflow2.py in scripts folder first

```python
import pandas as pd
from tqdm.auto import tqdm
import glob
```

```python
fs = glob.glob('../scripts/other-external/*')
dfs = []
for f in tqdm(fs):
    df = pd.read_csv(f)
    dfs.append(df)
df = pd.concat(dfs, ignore_index=True)
df
```

```python
import json
def transfer_prompt(x):
    x = json.loads(x)
    x = [xx for xx in x if xx is not None]
    x = ' '.join(x)
    return x
```

```python
df['prompt'] = df['prompt'].apply(transfer_prompt)
df['response_a'] = df['response_a'].apply(transfer_prompt)
df['response_b'] = df['response_b'].apply(transfer_prompt)
```

```python
df['author'] = df['id'].apply(lambda x: '-'.join(x.split('-')[:-1]))

```

```python
df.to_parquet('../data/2nd_place_external.parquet')
```
